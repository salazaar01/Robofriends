## **Marco Teórico**

### **1. Robots Seriales**
Un robot serial es una cadena cinemática abierta formada por eslabones conectados mediante juntas, que generalmente son rotacionales o prismáticas. El modelo URDF (Unified Robot Description Format) es un estándar en XML que permite describir la estructura física del robot, incluyendo parámetros geométricos, dinámicos y las relaciones entre eslabones y juntas. Este modelo es esencial para integrar el robot en entornos de simulación como Gazebo y ROS (Robot Operating System). En el caso de un robot serial 4R, este cuenta con cuatro grados de libertad (DoF) que le otorgan flexibilidad para realizar tareas específicas en el espacio tridimensional.

### **2. Cinemática Directa**
La cinemática directa se encarga de calcular la posición y orientación del efector final del robot (extremo del brazo robótico) a partir de los valores conocidos de las variables de las articulaciones (ángulos o desplazamientos). Este cálculo se realiza mediante la composición de matrices de transformación homogénea.

La matriz general de transformación homogénea para una articulación está dada por:
\[
T = 
\begin{bmatrix}
\cos\theta & -\sin\theta\cos\alpha & \sin\theta\sin\alpha & a\cos\theta \\
\sin\theta & \cos\theta\cos\alpha & -\cos\theta\sin\alpha & a\sin\theta \\
0 & \sin\alpha & \cos\alpha & d \\
0 & 0 & 0 & 1
\end{bmatrix}
\]
Donde:

- \\( \theta \\): Ángulo de la articulación (rotacional).
- \\( d \\): Desplazamiento lineal (prismático).
- \\( a \\): Longitud del eslabón.
- \\( \alpha \\): Ángulo entre los ejes \\( Z_i \\) y \\( Z_{i+1} \\).


### **3. Cinemática Inversa**
La cinemática inversa es el proceso de determinar los valores de las variables de las articulaciones necesarias para que el efector final alcance una posición y orientación deseada. Este problema es más complejo que la cinemática directa y, en algunos casos, puede tener múltiples soluciones o no tener solución alguna dependiendo de las restricciones del robot.

Para resolver la cinemática inversa, se parte de la relación entre las variables articulares y la posición del efector final, representada por la matriz de transformación homogénea:

\[
T = 
\begin{bmatrix}
R & p \\
0 & 1
\end{bmatrix}
\]

Donde:

- \\( R \\): Matriz de rotación 3 X 3 que define la orientación del efector final.
- \\( p = [x, y, z]^T \\): Vector de posición del efector final en coordenadas cartesianas.

En el caso del robot serial 4R, cada junta agrega un grado de libertad rotacional, por lo que el sistema de ecuaciones incluye las coordenadas finales y los ángulos articulares \\( \theta_1, \theta_2, \theta_3, \theta_4 \\).
### **4. Control de Trayectoria**
El control de trayectoria permite mover al robot de manera precisa a través de una ruta definida. Esto involucra el cálculo de puntos intermedios y la interpolación entre ellos para generar un movimiento suave. Los controladores de trayectoria suelen trabajar a nivel de juntas, ajustando los ángulos o desplazamientos lineales para cada articulación.

### **5. Simulación en Gazebo**
Gazebo es un entorno de simulación ampliamente utilizado en robótica, que permite probar y validar el diseño y control de robots en un entorno virtual. A través de modelos URDF y paquetes ROS, es posible representar robots, interactuar con ellos y aplicar algoritmos de control.

## Definición de las funciones

Matriz de transformaciones homógeneas


```python
from sympy import symbols, cos, sin, Matrix, pprint

# Definición de variables simbólicas
x_i_j, y_i_j, z_i_j, g_i_j, b_i_j, a_i_j = symbols('x_i_j y_i_j z_i_j g_i_j b_i_j a_i_j')

# Definición de la matriz de transformación homogénea
T_ij = Matrix([
    [cos(a_i_j)*cos(b_i_j), cos(a_i_j)*sin(b_i_j)*sin(g_i_j) - sin(a_i_j)*cos(g_i_j), cos(a_i_j)*sin(b_i_j)*cos(g_i_j) + sin(a_i_j)*sin(g_i_j), x_i_j],
    [sin(a_i_j)*cos(b_i_j), sin(a_i_j)*sin(b_i_j)*sin(g_i_j) + cos(a_i_j)*cos(g_i_j), sin(a_i_j)*sin(b_i_j)*cos(g_i_j) - cos(a_i_j)*sin(g_i_j), y_i_j],
    [-sin(b_i_j), cos(b_i_j)*sin(g_i_j), cos(b_i_j)*cos(g_i_j), z_i_j],
    [0, 0, 0, 1]
])

# Mostrar la matriz de transformación homogénea de manera bonita
pprint(T_ij)
```

Matriz homogénea genérica:

# **Reflexiones sobre las dificultades encontradas y cómo se resolvieron**.


    Dificultades con el modelo URDF

   
Una de las primeras dificultades encontradas fue la correcta especificación de las transformaciones entre eslabones en el modelo URDF, especialmente cuando se trataba de definir las posiciones relativas de las articulaciones y la orientación de los sistemas de referencia. El manejo de las coordenadas y la orientación de las juntas en relación a los sistemas de referencia del robot fue un desafío que se resolvió mediante una revisión minuciosa de la documentación de URDF y un enfoque paso a paso en la definición de cada componente del robot.

    Problemas con la cinemática inversa
   
La cinemática inversa del robot 4R presentó complejidades debido a las múltiples soluciones posibles y la necesidad de garantizar que las soluciones fueran físicamente viables. Para resolver esta dificultad, se utilizó una combinación de métodos algebraicos y numéricos, como la optimización, para restringir las soluciones a aquellas que cumplen con los límites de las articulaciones. A pesar de los esfuerzos, el cálculo de la cinemática inversa para configuraciones extremas sigue siendo un área que podría mejorar en futuras versiones del proyecto.

    Simulación en Gazebo y control de trayectoria
   
Durante la implementación del control de trayectoria, una dificultad relevante fue garantizar que el robot siguiera de manera fluida las trayectorias definidas sin colisionar con su propio cuerpo. Esto fue especialmente complejo al integrar los movimientos de las articulaciones en 3D, ya que se requería una sincronización precisa para evitar fallos. La solución pasó por ajustar los parámetros de los controladores y realizar pruebas de diferentes trayectorias hasta obtener un movimiento satisfactorio.

Además, incluir un sistema de verificación que identifique si las soluciones son físicamente posibles podría evitar movimientos erráticos o inviables.

    Integración de controladores avanzados

En el futuro, se podría investigar la integración de controladores más avanzados, como los controladores PID o el control adaptativo, para mejorar la precisión y la estabilidad en el seguimiento de trayectorias, especialmente en entornos con perturbaciones.

El modelo desarrollado hasta ahora está orientado principalmente a la simulación. Para llevar el proyecto a un nivel más avanzado, se podría trabajar en la implementación de este robot en hardware real, utilizando controladores físicos y sensores para validar la precisión del modelo en condiciones del mundo real. Esto implicaría la integración con plataformas de hardware como Arduino, Raspberry Pi o sistemas más avanzados de control en robótica industrial.

    Estudio de cinemática y dinámica más compleja
   
El análisis de la dinámica del robot, que incluye las fuerzas y torques en las articulaciones, es otro aspecto que se podría incluir en futuras extensiones del proyecto. A través de modelos dinámicos, sería posible simular no solo el movimiento del robot, sino también su comportamiento frente a cargas externas o impactos, lo cual es relevante para aplicaciones industriales.

# **Conclusiones generales**

Este proyecto ha logrado cumplir con su objetivo principal de desarrollar e implementar una simulación completa de un robot serial de 4 grados de libertad (4R) en un entorno tridimensional, aplicando los principios fundamentales de la robótica estudiados a lo largo del curso. A través de la creación de un modelo URDF detallado, la implementación de la cinemática directa e inversa, y la simulación en Gazebo, se ha podido generar un sistema funcional que no solo valida los conceptos teóricos, sino que también permite observar y controlar el comportamiento del robot en un entorno virtual. La integración de estos componentes ha demostrado la efectividad de los enfoques empleados, como la calibración precisa de los parámetros cinemáticos y el diseño de un controlador para la ejecución de trayectorias.

A lo largo del desarrollo del proyecto, se ha enfrentado a diversos retos técnicos que han enriquecido la experiencia de aprendizaje. La resolución de la cinemática inversa, aunque compleja, ha sido alcanzada mediante el uso de algoritmos de optimización, lo que ha permitido superar las dificultades inherentes al cálculo de las configuraciones articulares del robot. La simulación en Gazebo, por su parte, ha servido como una plataforma crucial para verificar la precisión del modelo y ajustar los parámetros de control, permitiendo evaluar en tiempo real el desempeño del robot en un espacio tridimensional.

Este trabajo no solo ha permitido consolidar una comprensión más profunda de los aspectos teóricos de la robótica, sino que también ha proporcionado una base sólida para futuras investigaciones y desarrollos en el campo. Las herramientas utilizadas, como URDF, Gazebo y ROS, han demostrado ser esenciales para la simulación y control de robots, y su aplicación en proyectos más complejos, como la integración de robots con más grados de libertad o la interacción con entornos físicos, abre una amplia gama de posibilidades para continuar expandiendo este tipo de investigaciones.
